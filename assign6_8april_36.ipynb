```python
x=[4,5,10,4,3,11,14,6,10,12]
y=[21,19,24,17,16,25,24,22,21,21]
z=[34,67,89,65,54,32,21,23,24,25]
data=list(zip(x,y,z))
print(data)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df=pd.read_csv('/content/data setmam.csv')
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
plt.scatter(x,y,c=kmeans.labels_)
plt.show()



```

```python
plt.scatter(df['Duration'], df['Pulse'], c=kmeans.labels_)

plt.show()
```
